```python
import pandas as pd
import seaborn as sns
import numpy as np
import scipy as sp
import sklearn
```

```python
combined = pd.read_csv("../feature_store/feature_data_milestone_1", index_col="PatientID")
```

```python
combined.head()
```

```python
# Determine correlations using the target variable

corrs = combined.corr(numeric_only=True)["outcome"]
```

```python
# Use all features

X = combined[corrs.index]
```

```python
X.head()
```

```python
# separate the target variable

y = combined["outcome"]

X = X.drop(columns=["outcome", "birth_year"])
```

```python
# glucose_insulin_difference is experimental
# may need to remove this feature depending
X.describe()
```

```python
# Split data into 80/20 train/test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=23, test_size=.20)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV


from sklearn.pipeline import Pipeline

scaler_pipeline = Pipeline(steps=[("scaler", RobustScaler()), ("poly", PolynomialFeatures(degree=2))])


preprocessor = ColumnTransformer(
    transformers=[
        ("scaler", scaler_pipeline, X.columns)
    ]
)

param_grid = [
    {
        "solver": ["newton-cg", "sag", "lbfgs"]
    },
    {
        "solver": ["liblinear"],
        "penalty": ["l1", "l2"]
    },
    {
        "solver": ["saga"],
        "penalty": ["l1", "l2"],
        
    },
    {
        "solver": ["saga"],
        "penalty": ["elasticnet"],
        "l1_ratio": [0.0, 0.5, 1.0]
    }
]

grid = GridSearchCV(LogisticRegression(max_iter=10000), param_grid, cv=5)

lr_estimator = Pipeline(steps=[("preprocessor", preprocessor), ("model", grid)])

lr_estimator.fit(X_train, y_train)

```

```python
# Check the raw accuracy

lr_estimator.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_predictions = lr_estimator.predict(X_test)
```

```python
# Check the confusion matrix
from sklearn.metrics import roc_auc_score

cm = confusion_matrix(y_test, y_predictions)
```

```python
roc_auc_score(y_test, y_predictions)
```

```python
# True negatives, false positives
# False negatives, true positives

cm
```

```python
cm_graphic = ConfusionMatrixDisplay(confusion_matrix=cm)
```

```python
cm_graphic.plot()
```

```python
from sklearn.metrics import classification_report

# Precision -accuracy of the positive predictions
# Recall - sensitivity to the true positive rate
# f1 score - weighted harmonic mean between 1 (best) 
# and 0 (worst)
# support - number of instances of class in 
# test (true) data


print(classification_report(y_test, y_predictions))
```

```python
from sklearn.model_selection import TunedThresholdClassifierCV

tuned_classifier = TunedThresholdClassifierCV(lr_estimator, scoring="balanced_accuracy")
```

```python
tuned_classifier.fit(X_train, y_train)
```

```python
tuned_y_preds = tuned_classifier.predict(X_test)
```

```python
y_predictions = lr_estimator.predict(X_test)
```

```python
roc_auc_score(y_test, tuned_y_preds)
```

```python
print(classification_report(y_test, tuned_y_preds))
```
